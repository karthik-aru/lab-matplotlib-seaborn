#### Use your previous code but now, resize your plot.

```python
# your code here
fig2, ax2 = plt.subplots(1,2, figsize=(16,6))
fig2.suptitle("Line plots split in two subplots", fontsize=20, fontweight="bold")
ax2[0].plot(x, y, label="2x", linewidth=2, linestyle="-.", c="red")
ax2[0].set_title("Linear")
ax2[1].plot(x, z, label="x^2", linewidth=2, linestyle=":", c="violet")
ax2[1].set_title("Exponential")

for ax in ax2:
    ax.legend()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
month_wise = fitbit.groupby(["Months_encoded"]).aggregate({"Steps":"median"}).reset_index()

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
week_wise = fitbit.groupby(["Work_or_Weekend"]).aggregate({"Steps":"median"}).reset_index()
```

```python
month_wise
```

```python
figm, axm = plt.subplots(figsize=(12,4))
monthfig = sns.lineplot(data=month_wise, x="Months_encoded", y="Steps")
axm.set(xlabel="Months")
monthfig = monthfig.get_figure()
monthfig.savefig("../plots/fitbit/month_wise.png")
```

```python
figw, axw = plt.subplots(figsize=(5,5))
fig_week = sns.barplot(data=week_wise, x="Work_or_Weekend", y="Steps")
axw.set_xticklabels(["Workday", "Weekend"])
fig_week = fig_week.get_figure()
fig_week.savefig("../plots/fitbit/week_wise.png")
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
fitbit.head()
minutes_v_steps = fitbit[["Minutes Lightly Active", "Minutes Very Active", "Minutes Sedentary", "Steps"]].copy()
m_v_s_colnames = list(minutes_v_steps.columns)
figms, axms = plt.subplots(1,3, figsize=(16,7))
figms.suptitle("Scatterplot of Minute-wise-activity and steps taken", fontsize=20)
for i in range(3):
    axms[i].scatter(x=minutes_v_steps[m_v_s_colnames[i]], y=minutes_v_steps["Steps"])
    axms[i].set_title(m_v_s_colnames[i] + " vs Steps")
    axms[i].set_xlabel(m_v_s_colnames[i])
    axms[i].set_ylabel("Steps")

figms.savefig("../plots/fitbit/scatter_minute_wise.png")
```
